```python

# ============================================
# Zero-Knowledge Proof - Schnorr Protocol Lab
# Special Topics in Information Security
# Seiyun University
# ============================================

#@title **ุงูุฌุฒุก 1: ุงุณุชูุฑุงุฏ ุงูููุชุจุงุช**
import hashlib
import random
from sympy import isprime, randprime

print("=" * 60)
print("ูุฎุชุจุฑ ุจุฑูุชูููู ุดููุฑ ูุฅุซุจุงุช ุงููุนุฑูุฉ ุตูุฑ (ZKP)")
print("=" * 60)
print()

#@title **ุงูุฌุฒุก 2: ุชูููุฏ ุงููุนุงููุงุช ุงูุขููุฉ**
def generate_parameters(bit_length=8):
    """
    ุชูููุฏ ุงูุฃุนุฏุงุฏ ุงูุฃูููุฉ (p) ูุงูุฌุฐุฑ ุงูุจุฏุงุฆู (g)
    bit_length: ุทูู ุงูุจุชุงุช (ุตุบูุฑ ููุชูุถูุญ)
    """
    # ุชูููุฏ ุนุฏุฏ ุฃููู p ูุจูุฑ ุจูุง ูููู
    p = randprime(2**(bit_length-1), 2**bit_length)

    # ุงูุจุญุซ ุนู ุฌุฐุฑ ุจุฏุงุฆู g (ููุชุจุณูุทุ ูุณุชุฎุฏู 2 ุฅุฐุง ูุงู ุฌุฐุฑุงู ุจุฏุงุฆูุงู)
    # ูู ุงูุชุทุจูู ุงูุญููููุ ูุฌุจ ุงูุชุญูู ูู ุฃู g ูู ุจุงููุนู ุฌุฐุฑ ุจุฏุงุฆู ููุฌููุนุฉ p
    g = 2
    while pow(g, (p-1)//2, p) == 1:  # ุงุฎุชุจุงุฑ ูุจุณุท
        g += 1

    return p, g

# ุชูููุฏ ุงููุนุงููุงุช
p, g = generate_parameters(10)  # 10 ุจุชุงุช ููุชูุถูุญ
print("[1] ุชูููุฏ ุงููุนุงููุงุช:")
print(f"   ุงูุนุฏุฏ ุงูุฃููู (p) = {p}")
print(f"   ุงูุฌุฐุฑ ุงูุจุฏุงุฆู (g) = {g}")
print()

#@title **ุงูุฌุฒุก 3: ุชูููุฏ ุงูููุงุชูุญ (ุงูุณุฑ ุงูุฎุงุต ูุงูููุชุงุญ ุงูุนุงู)**
# ุงูุณุฑ ุงูุณุฑู ูููุซุจุช (Prover)
x = random.randint(2, p-2)  # ุงูููุชุงุญ ุงูุฎุงุต
y = pow(g, x, p)            # ุงูููุชุงุญ ุงูุนุงู: y = g^x mod p

print("[2] ุชูููุฏ ุงูููุงุชูุญ:")
print(f"   ุงูููุชุงุญ ุงูุณุฑู ูููุซุจุช (x) = {x}")
print(f"   ุงูููุชุงุญ ุงูุนุงู (y = g^x mod p) = {y}")
print()

#@title **ุงูุฌุฒุก 4: ุงูุชูููุฐ ุงูุตุงุฏู ููุจุฑูุชูููู ุงูุชูุงุนูู**
def honest_schnorr_protocol(p, g, x, y, verbose=True):
    """
    ุชูููุฐ ุจุฑูุชูููู ุดููุฑ ุงูุตุงุฏู (ุงููุซุจุช ูุนุฑู ุงูุณุฑ x)
    """
    if verbose:
        print("[3] ุชุดุบูู ุจุฑูุชูููู ุดููุฑ ุงูุตุงุฏู (ุงูุชูุงุนูู):")
        print("   " + "-" * 40)

    # ุงูุฎุทูุฉ 1: ุงููุซุจุช ูุฎุชุงุฑ ุนุดูุงุฆูุงู r ููุญุณุจ t = g^r mod p
    r = random.randint(1, p-1)
    t = pow(g, r, p)
    if verbose:
        print(f"   1. ุงููุซุจุช ูุฎุชุงุฑ r = {r}")
        print(f"      ููุญุณุจ ุงูุงูุชุฒุงู t = g^r mod p = {t}")

    # ุงูุฎุทูุฉ 2: ุงููุฏูู ูููุฏ ุชุญุฏู ุนุดูุงุฆู e
    e = random.randint(0, 1)  # ูุณุชุฎุฏู 0 ุฃู 1 ููุชุจุณูุท
    if verbose:
        print(f"   2. ุงููุฏูู ูููุฏ ุชุญุฏู ุนุดูุงุฆู e = {e}")

    # ุงูุฎุทูุฉ 3: ุงููุซุจุช ูุญุณุจ ุงูุงุณุชุฌุงุจุฉ s = r + e*x mod (p-1)
    s = (r + e * x) % (p-1)
    if verbose:
        print(f"   3. ุงููุซุจุช ูุญุณุจ ุงูุงุณุชุฌุงุจุฉ s = r + e*x = {s}")

    # ุงูุฎุทูุฉ 4: ุงููุฏูู ูุชุญูู ูู ุฃู g^s โก t * y^e (mod p)
    left_side = pow(g, s, p)
    right_side = (t * pow(y, e, p)) % p

    verification_passed = (left_side == right_side)

    if verbose:
        print(f"   4. ุงููุฏูู ูุชุญูู: g^s mod p = {left_side}")
        print(f"                 t * y^e mod p = {right_side}")
        print(f"      ุงููุชูุฌุฉ: {'ูุฌุงุญ โ' if verification_passed else 'ูุดู โ'}")
        print()

    return verification_passed, t, e, s

# ุชุดุบูู ุงูุจุฑูุชูููู ุงูุตุงุฏู
print("=" * 60)
verification_passed, t, e, s = honest_schnorr_protocol(p, g, x, y)

#@title **ุงูุฌุฒุก 5: ูุญุงููุฉ ุงูุบุด (ูุซุจุช ูุง ูุนุฑู ุงูุณุฑ)**
def cheating_prover(p, g, y, verbose=True):
    """
    ูุญุงููุฉ ุงูุบุด: ูุซุจุช ูุง ูุนุฑู ุงูุณุฑ x ูุญุงูู ุฎุฏุงุน ุงููุฏูู
    """
    if verbose:
        print("[4] ูุญุงููุฉ ุงูุบุด (ูุซุจุช ูุง ูุนุฑู ุงูุณุฑ x):")
        print("   " + "-" * 40)

    # ุงุณุชุฑุงุชูุฌูุฉ ุงูุบุด: ุชุฎููู ุงูุชุญุฏู ูุณุจูุงู
    # ุงูุทุฑููุฉ 1: ุฅุฐุง ุฎูู e=0 ุจุดูู ุตุญูุญุ ููููู ุงุณุชุฎุฏุงู ุฃู r
    # ุงูุทุฑููุฉ 2: ุฅุฐุง ุฎูู e=1ุ ูุฌุจ ุฃู ูุญุณุจ t = g^s / y

    guessed_e = random.randint(0, 1)  # ูุฎูู ุงูุชุญุฏู

    if guessed_e == 0:
        # ุฅุฐุง ูุงู ุงูุชุญุฏู e=0ุ ูุงูุงุณุชุฌุงุจุฉ ูุง ุชุนุชูุฏ ุนูู x
        r_fake = random.randint(1, p-1)
        t_fake = pow(g, r_fake, p)
        s_fake = r_fake
        if verbose:
            print(f"   ุงููุซุจุช ุงููุฎุงุฏุน ูุฎูู ุฃู ุงูุชุญุฏู ุณูููู e = 0")
            print(f"   ูุฎุชุงุฑ r = {r_fake}, t = {t_fake}, s = {s_fake}")
    else:
        # ุฅุฐุง ูุงู ุงูุชุญุฏู e=1ุ ูุญุงูู ุงุฎุชูุงุฑ s ุฃููุงู ุซู ูุญุณุจ t
        s_fake = random.randint(1, p-1)
        # t = g^s / y mod p = g^s * y^(-1) mod p
        y_inv = pow(y, p-2, p)  # ูุนููุณ y ุจุงุณุชุฎุฏุงู ูุจุฑููุฉ ููุฑูุง ุงูุตุบุฑู
        t_fake = (pow(g, s_fake, p) * y_inv) % p
        if verbose:
            print(f"   ุงููุซุจุช ุงููุฎุงุฏุน ูุฎูู ุฃู ุงูุชุญุฏู ุณูููู e = 1")
            print(f"   ูุฎุชุงุฑ s = {s_fake}, ุซู ูุญุณุจ t = {t_fake}")

    # ุงููุฏูู ูููุฏ ุงูุชุญุฏู ุงูุญูููู (ุจุดูู ุนุดูุงุฆู)
    actual_e = random.randint(0, 1)

    # ุงูุชุญูู
    if actual_e == guessed_e:
        # ุฅุฐุง ูุงู ุชุฎูููู ุตุญูุญุงูุ ูุงูุชุตุฏูู ุณููุฌุญ
        left_side = pow(g, s_fake, p)
        if guessed_e == 0:
            right_side = t_fake % p
        else:
            right_side = (t_fake * pow(y, 1, p)) % p

        verification_passed = (left_side == right_side)
        if verbose:
            print(f"   ุงูุชุญุฏู ุงููุนูู ูู ุงููุฏูู: e = {actual_e}")
            print(f"   ุชุฎููู ุงููุฎุงุฏุน ูุงู {'ุตุญูุญ โ' if actual_e == guessed_e else 'ุฎุงุทุฆ โ'}")
            print(f"   ูุชูุฌุฉ ุงูุชุญูู: {'ูุฌุงุญ โ' if verification_passed else 'ูุดู โ'}")
    else:
        verification_passed = False
        if verbose:
            print(f"   ุงูุชุญุฏู ุงููุนูู ูู ุงููุฏูู: e = {actual_e}")
            print(f"   ุชุฎููู ุงููุฎุงุฏุน ูุงู {'ุตุญูุญ โ' if actual_e == guessed_e else 'ุฎุงุทุฆ โ'}")
            print(f"   ูุชูุฌุฉ ุงูุชุญูู: ูุดู โ (ุงูุชุฎููู ุฎุงุทุฆ)")

    print()
    return verification_passed

# ุชุดุบูู ูุญุงููุฉ ุงูุบุด
cheating_result = cheating_prover(p, g, y)

#@title **ุงูุฌุฒุก 6: ููุงุณ ุงุญุชูุงู ุงููุฌุงุญ ูู ุงูุบุด**
def measure_cheating_probability(p, g, y, num_trials=100):
    """
    ุญุณุงุจ ุงุญุชูุงู ูุฌุงุญ ุงููุฎุงุฏุน ุจุนุฏ ุนุฏุฉ ูุญุงููุงุช
    """
    print("[5] ุชุฌุฑุจุฉ ููุงุณ ุงุญุชูุงู ุงููุฌุงุญ ูู ุงูุบุด:")
    print("   " + "-" * 40)

    successes = 0
    for i in range(num_trials):
        if cheating_prover(p, g, y, verbose=False):
            successes += 1

    probability = successes / num_trials
    print(f"   ุนุฏุฏ ุงููุญุงููุงุช: {num_trials}")
    print(f"   ุนุฏุฏ ูุฑุงุช ุงููุฌุงุญ: {successes}")
    print(f"   ุงุญุชูุงู ุงููุฌุงุญ ูู ุงูุบุด: {probability:.3f} ({probability*100:.1f}%)")
    print(f"   ุงูุงุญุชูุงู ุงููุธุฑู ุงููุชููุน: 0.5 (50%)")
    print()
    return probability

# ููุงุณ ุงุญุชูุงู ุงูุบุด
cheat_prob = measure_cheating_probability(p, g, y, 50)

#@title **ุงูุฌุฒุก 7: ุชุญููู ููุงุช-ุดุงููุฑ (ุบูุฑ ุชูุงุนูู)**
def fiat_shamir_non_interactive(p, g, x, y, message="HelloZKP", verbose=True):
    """
    ุชุญููู ููุงุช-ุดุงููุฑ: ุจุฑูุชูููู ุบูุฑ ุชูุงุนูู
    """
    if verbose:
        print("[6] ุชุญููู ููุงุช-ุดุงููุฑ (ุจุฑูุชูููู ุบูุฑ ุชูุงุนูู):")
        print("   " + "-" * 40)

    # ุงูุฎุทูุฉ 1: ุงููุซุจุช ูุฎุชุงุฑ r ุนุดูุงุฆูุงู ููุญุณุจ t = g^r mod p
    r = random.randint(1, p-1)
    t = pow(g, r, p)
    if verbose:
        print(f"   1. ุงููุซุจุช ูุฎุชุงุฑ r = {r}")
        print(f"      ููุญุณุจ t = g^r mod p = {t}")

    # ุงูุฎุทูุฉ 2: ุชูููุฏ ุงูุชุญุฏู ุจุงุณุชุฎุฏุงู ุฏุงูุฉ ุงููุงุด (ุจุฏูุงู ูู ุงููุฏูู)
    # e = H(t || message)
    hash_input = f"{t}{message}".encode()
    e_hash = hashlib.sha256(hash_input).hexdigest()
    e = int(e_hash, 16) % 2  # ูุฃุฎุฐ ุจุชุฉ ูุงุญุฏุฉ ููุชุจุณูุท

    if verbose:
        print(f"   2. ูุญุณุจ ุงูุชุญุฏู e = H(t || message)")
        print(f"      t = {t}, message = '{message}'")
        print(f"      SHA256 hash = {e_hash[:16]}...")
        print(f"      e ุงููุณุชุฎุฑุฌ = {e}")

    # ุงูุฎุทูุฉ 3: ุงููุซุจุช ูุญุณุจ s = r + e*x mod (p-1)
    s = (r + e * x) % (p-1)

    if verbose:
        print(f"   3. ูุญุณุจ ุงูุงุณุชุฌุงุจุฉ s = r + e*x = {s}")

    # ุงูุฎุทูุฉ 4: ุฃู ูุฏูู ููููู ุงูุชุญูู ุจุงุณุชุฎุฏุงู tุ eุ s
    # ุฃููุงู: ุฅุนุงุฏุฉ ุญุณุงุจ e ุจููุณ ุงูุทุฑููุฉ
    hash_input_verify = f"{t}{message}".encode()
    e_verify_hash = hashlib.sha256(hash_input_verify).hexdigest()
    e_verify = int(e_verify_hash, 16) % 2

    # ุซู ุงูุชุญูู ูู g^s โก t * y^e (mod p)
    left_side = pow(g, s, p)
    right_side = (t * pow(y, e_verify, p)) % p

    verification_passed = (left_side == right_side)

    if verbose:
        print(f"   4. ุงูุชุญูู:")
        print(f"      - ุฅุนุงุฏุฉ ุญุณุงุจ e ูู t ูุงูุฑุณุงูุฉ = {e_verify}")
        print(f"      - ุญุณุงุจ g^s mod p = {left_side}")
        print(f"      - ุญุณุงุจ t * y^e mod p = {right_side}")
        print(f"      ุงููุชูุฌุฉ: {'ูุฌุงุญ โ' if verification_passed else 'ูุดู โ'}")
        print()

    return verification_passed, t, e, s

# ุชุดุบูู ุงููุณุฎุฉ ุบูุฑ ุงูุชูุงุนููุฉ
print("=" * 60)
ni_passed, ni_t, ni_e, ni_s = fiat_shamir_non_interactive(p, g, x, y)

#@title **ุงูุฌุฒุก 8: ุชุณุฌูู ุงููุชุงุฆุฌ**
print("=" * 60)
print("[7] ููุฎุต ุงููุชุงุฆุฌ:")
print("   " + "-" * 40)

print("   1. ุงูุจุฑูุชูููู ุงูุตุงุฏู (ุงูุชูุงุนูู):")
print(f"      - ุงูุงูุชุฒุงู t = {t}")
print(f"      - ุงูุชุญุฏู e = {e}")
print(f"      - ุงูุงุณุชุฌุงุจุฉ s = {s}")
print(f"      - ุงูุชุญูู: {'ูุฌุญ' if verification_passed else 'ูุดู'}")

print("   2. ูุญุงููุฉ ุงูุบุด:")
print(f"      - ุงููุชูุฌุฉ: {'ูุฌุญ ุงููุฎุงุฏุน' if cheating_result else 'ูุดู ุงููุฎุงุฏุน'}")

print("   3. ุชุญููู ููุงุช-ุดุงููุฑ (ุบูุฑ ุชูุงุนูู):")
print(f"      - t = {ni_t}, e = {ni_e}, s = {ni_s}")
print(f"      - ุงูุชุญูู: {'ูุฌุญ' if ni_passed else 'ูุดู'}")

print("   4. ุชุฌุฑุจุฉ ุงุญุชูุงู ุงูุบุด:")
print(f"      - ุงูุงุญุชูุงู ุงูููุงุณ: {cheat_prob:.3f}")
print(f"      - ุงูุชูุณูุฑ: ูููุฎุงุฏุน ูุฑุตุฉ 50% ูุชุฎููู ุงูุชุญุฏู ุงูุตุญูุญ")
print(f"                 (ูุฃู e ุชุฃุฎุฐ ููู 0 ุฃู 1 ูู ูุซุงููุง ุงููุจุณุท)")

print("=" * 60)
print("ุชู ุงูุงูุชูุงุก ูู ุชูููุฐ ูุฎุชุจุฑ ุจุฑูุชูููู ุดููุฑ! ๐ฏ")
print("=" * 60)
```
